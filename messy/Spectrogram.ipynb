```python
def collect_target_and_featers(timesequence, n_fft=100, target_len=20, spec_len=7):
    spectrogram = np.log(np.abs(librosa.stft(timesequence, n_fft=n_fft, center=False))**2)
    res = []
    time_target_start = n_fft + spec_len * n_fft // 4
    for i in range(spectrogram.shape[1] - spec_len - 1):
        featers = spectrogram[:, i: i + spec_len]
        target = timesequence[time_target_start : time_target_start + target_len]
        time_target_start += n_fft // 4
        res.append((featers, target))
    return res    
```

```python
class MyDataset(Dataset):
    def __init__(self, data):
        self.data = data
             
    def __getitem__(self, index):
        x = self.data[index][0]
        y = self.data[index][1]
        
        return x, y
    
    def __len__(self):
        return len(self.data)
```

```python
dataset = MyDataset(dataset)
```

```python
train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True, drop_last=True)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False, drop_last=True)
```

```python
class SpecModel(nn.Module):
    def __init__(self):
        super(SpecModel, self).__init__()
        ################################################################################
        self.L1 = torch.nn.Linear(51, 51)
        self.relu = torch.nn.ReLU()
        self.lstm = nn.LSTM(input_size=51, hidden_size=51, num_layers=1, batch_first=True, bidirectional=True)
        #self.gru = torch.nn.GRU(input_size = 64, hidden_size = 28, batch_first = True, bidirectional = True)
        #self.pack = torch.nn.utils.rnn.pack_padded_sequence(input, lengths, batch_first=False, enforce_sorted=True)
        self.L2 = torch.nn.Linear(in_features = 2*51, out_features=1)
        ################################################################################

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        
        ################################################################################
        x = x.transpose(1, 2)
        x = self.L1(x)
        x = self.relu(x)
        #x = torch.nn.utils.rnn.pack_padded_sequence(x, lengths, batch_first = True)
        #x = self.gru(x)[0].data
        
        output, (ht, ct) = self.lstm(x)
        x = torch.tanh(ht.transpose(0, 1))
        #print(output.size())
        x = torch.nn.Flatten()(x)
        x = self.L2(x)
        ################################################################################
        return x
```

```python
model = SpecModel().to('cuda')
```

```python
def run_epoch(model, dataloader, is_training=False):
    epoch_loss = 0

    if is_training:
        model.train()
    else:
        model.eval()

    for idx, (x, y) in enumerate(dataloader):
        if is_training:
            optimizer.zero_grad()

        batchsize = x.shape[0]
        #print(x.shape)
        #x = torch.FloatTensor(x)
        #y = torch.LongTensor(y)
        
        x = x.float().to('cuda')

        y = y.view(-1).float().to('cuda')

        out = model(x)
        loss = criterion(out.contiguous(), y.contiguous())

        if is_training:
            loss.backward()
            optimizer.step()

        epoch_loss += (loss.detach().item() / batchsize)


    return epoch_loss
```

```python
# define optimizer and loss function
criterion = nn.MSELoss()
lr = 0.0001
epochs = 5

losses_train = []
losses_test = []
# begin training
for epoch in range(epochs): 
#     if epoch % 100 == 0:
#         lr *= 0.9
    optimizer = optim.Adam(model.parameters(), lr=lr, betas=(0.9, 0.98), eps=1e-9)    
    loss_train = run_epoch(model, train_dataloader, is_training=True)
    loss_val = run_epoch(model, test_dataloader)
    losses_train.append(loss_train)
    losses_test.append(loss_val)
    
    clear_output(True)
    fig = plt.figure(figsize=(10, 9))
    
    ax_1 = fig.add_subplot(2, 1, 1)
    ax_2 = fig.add_subplot(2, 1, 2)
    ax_1.set_title('train')
    ax_1.plot(losses_train)
    ax_2.set_title('test')
    ax_2.plot(losses_test)
    plt.show()
    
    print('Epoch[{}/{}] | loss train:{:.6f}, test:{:.6f}'
              .format(epoch+1, epochs, loss_train, loss_val))
```

```python
mean = np.mean(a[:, 1])
std = np.std(a[:, 1])
Hs = mean + 4*std
```

```python
(data > Hs).sum()
```

```python
cnt = 0
i = 20
idxs = []
sums = []
while i < len(data) - 20:
    if data[i] > Hs and (data[i: i + 50] > Hs).sum() > 1:
        idxs.append(i)
        sums.append((data[i: i + 50] > Hs).sum() - 1)
        i += 50
        cnt += 1
        
    i += 1
cnt
```

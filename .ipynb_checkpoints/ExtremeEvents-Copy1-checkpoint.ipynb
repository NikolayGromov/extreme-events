```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
```

```python
rho = 28.0
sigma = 10.0
beta = 8.0 / 3.0

def f(state, t):
    x, y, z = state  # unpack the state vector
    return sigma * (y - x), x * (rho - z) - y, x * y - beta * z  # derivatives

state0 = [1.0, 1.0, 1.0]
t = np.arange(0.0, 5.0, 0.01)

states = odeint(f, state0, t)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(states[:,0], states[:,1], states[:,2])
```

```python
np.random.seed(42)
data = []
target = []
for i in range(10000):
    state0 = [np.random.normal(1, 2), np.random.normal(1, 2), np.random.normal(1, 2)]
    states = odeint(f, state0, t)
    data.append(states[0])
    target.append(states[-1])
    
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import torchvision, torchvision.transforms as transforms
```

```python
dataset = [(data[i], target[i]) for i in range(len(data))]
```

```python
len(dataset)
```

```python
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [8000, 2000],
                                                            generator=torch.Generator().manual_seed(42))
```

```python
train_dataset[0]
```

```python
batch_size=16
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=batch_size, shuffle=True)
```

```python
model = nn.Sequential()

model.add_module("linear1", nn.Linear(3, 3))
model.add_module("relu", nn.ReLU())
model.add_module("linear2", nn.Linear(3, 10))
model.add_module("relu", nn.ReLU())
model.add_module("linear3", nn.Linear(10, 3))


model.cuda()
```

```python
from IPython.display import clear_output
```

```python
train_loss = []
val_loss = []
val_accuracy_for_print = []
train_loss_for_print = []

criterion = torch.nn.MSELoss() 
opt = torch.optim.Adam(model.parameters(),lr=0.001)

num_epochs = 100 # total amount of full passes over training data
for epoch in range(num_epochs):
    #start_time = time.time()
    model.train(True) # enable dropout / batch_norm training behavior
    for (X_batch, y_batch) in train_loader:
        # train on batch
        X_batch = X_batch.to(device='cuda', dtype=torch.float32)
        y_batch = y_batch.to(device='cuda', dtype=torch.float32)
        pred = model(X_batch)
        loss = criterion(pred, y_batch)
        loss.backward()
        opt.step()
        opt.zero_grad()
        train_loss.append(loss.cpu().data.numpy())
    
    model.train(False) # disable dropout / use averages for batch_norm
    for X_batch, y_batch in val_loader:
        X_batch = X_batch.to(device='cuda', dtype=torch.float32)
        y_batch = y_batch.to(device='cuda', dtype=torch.float32)
        pred = model(X_batch)
        loss_val = criterion(pred, y_batch)
        val_loss.append(loss_val.cpu().data.numpy())
    
    #if epoch == 7:
        #opt = torch.optim.Adam(model.parameters(),lr=0.0001)

    
    # Then we print the results for this epoch:
    val_accuracy_for_print.append(np.mean(val_loss[-len(val_dataset) // batch_size :]))
    train_loss_for_print.append(np.mean(train_loss[-len(train_dataset) // batch_size :]))
    clear_output(True)
    plt.figure(figsize=[16, 9])

    plt.subplot(2, 1, 1)
    plt.title("loss on train")
    plt.plot(train_loss_for_print)
            
    plt.subplot(2, 1, 2)
    plt.title("validation loss per epoch")
    plt.plot(val_accuracy_for_print)
            



    plt.show()
```

```python
i = 0
prediction = np.empty((2000, 3))
Y_test = np.empty((2000, 3))
for X_batch, y_batch in val_loader:
        X_batch = X_batch.to(device='cuda', dtype=torch.float32)
        y_batch = y_batch.to(device='cuda', dtype=torch.float32)
        pred = model(X_batch).cpu().data.numpy()
        prediction[i*16:(i+1)*16] = pred
        y_batch = y_batch.cpu().data.numpy()
        Y_test[i*16:(i+1)*16] = y_batch
        i += 1
        #loss_val = criterion(pred, y_batch)
        #val_loss.append(loss_val.cpu().data.numpy())
```

```python
from scipy.integrate import solve_ivp
```

```python
a = -0.025794
c = 0.02
b1 = 0.0135
b2 = 0.0065
k = 0.129
sigma = 1
alpha = 0.35
eps = 1
beta = 10e-8
q=1.6
```

```python
def FHN(self, X):
        x1, y1, x2, y2 = X
        return [x1*(a-x1)*(x1-1) - y1 + k*(x2-x1),
                b1*x1 - c*y1,
                x2*(a-x2)*(x2-1) - y2 + k*(x1-x2),
                b2*x2 - c*y2
               ]
```

```python
sol = solve_ivp(FHN, [0, 10000], [-0.1, 0.1, 0.1, 0.1], rtol = 1e-11, atol = 1e-11, dense_output=True)
```

```python
plt.plot(sol.t,sol.y[0])
```

```python
train = sol.y[0][:21600]
test = sol.y[0][21600:] 
```

```python
W_in = np.random.uniform(-sigma, sigma, (1000, 1))
```

```python
from skimage.io import imshow
```

```python
def erdos_renyi(n=1000, p=0.6):
    A = np.zeros((n, n))
    for i in range(n):
        for j in range(i, n):
            if np.random.uniform(0, 1) > p:
                a = np.random.uniform(-1, 1)
                A[i][j] = a
                A[j][i] = a
                
        A[i][i] = 0
            
    return A
```

```python
def reservoir(U):
    r = np.random.uniform(1, -1, (1000, 1))
    R = np.zeros((1000, len(U)))
    R[:, 0] = r[:, 0]
    for i in range(1, len(U)):
        r = (1-alpha)*r + np.tanh(A@r + W_in*U[i] + eps) #U[i] is a number
        R[:, i] = r[:, 0]
    return R
```

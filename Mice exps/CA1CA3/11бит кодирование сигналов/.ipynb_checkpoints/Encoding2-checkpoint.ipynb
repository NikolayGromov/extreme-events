```python
conv_ca3_300 = np.clip((-ca3_300[5000-1024:7048+1024] - MIN), a_min=0, a_max=None)  # *1.25
conv_ca3_1000 = np.clip((-ca3_1000[5000-1024:7048+1024] - MIN), a_min=0, a_max=None)
conv_ca1_300 = np.clip((-ca1_300[5000-1024:7048+1024] - MIN), a_min=0, a_max=None)
conv_ca1_1000 = np.clip((-ca1_1000[5000-1024:7048+1024] - MIN), a_min=0, a_max=None)
```

```python
plt.plot(-ca3_300[5000:7048] - MIN)
```

```python
plt.plot(conv_ca3_300)
```

```python
def Convert(x, old_low=0, old_high=3.3, new_low=0, new_high=4095):
    interval_0_1 = (x - old_low) / (old_high - old_low)
    scaled = new_low + (new_high - new_low) * interval_0_1
    return scaled.astype(int)
    
```

```python
conv_ca3_300 = Convert(conv_ca3_300)
conv_ca3_1000 = Convert(conv_ca3_1000)
conv_ca1_300 = Convert(conv_ca1_300)
conv_ca1_1000 = Convert(conv_ca1_1000)
```

```python
def FullReadConvertSave(filename):
    
    str_amp = filename[-7:-4]
    if filename[-8].isdigit():
        str_amp = filename[-8] + str_amp
    amp = np.load(filename)
    amp = np.concatenate((amp[:, 0], np.zeros(1096)))
    fig, ax = plt.subplots(nrows=2)
    ax[0].plot(amp)
    print("amp size", amp.shape)
    print("MAX", (-amp).max())
    #MIN = (-amp[5000-1024:7048+1024]).min()

    amp = np.clip(-amp, a_min=0, a_max=None)  # *1.25

    amp = Convert(amp)
    
    ax[1].plot(amp)

    np.savetxt("Converted_" + str_amp + "_4096.txt", amp)
    
    
```

```python
FullReadConvertSave("amp_100.npy")
```

```python
FullReadConvertSave("amp_200.npy")
FullReadConvertSave("amp_300.npy")
FullReadConvertSave("amp_400.npy")
FullReadConvertSave("amp_500.npy")
FullReadConvertSave("amp_1000.npy")
```

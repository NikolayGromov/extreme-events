```python
import pandas as pd
import numpy as np
import os
```

```python
path = "CrossValidationResults/"
```

```python
df = pd.DataFrame(columns=['name', 'precision', 'recall', "f1"])
```

```python
i = 0
for exp_name in os.listdir(path):
    if "SEED6Records" not in exp_name and "Slumber" not in exp_name and "ipynb" not in exp_name:
        precision = []
        recall = []
        f1 = []
        for split in os.listdir(path+exp_name):
            if "ipynb" not in split:
                f = open(path + exp_name + "/" + split + "/Metrics.txt", 'r')
                text = f.read()

                precision.append(float(text.split("\n")[1].split(" ")[1]))
                recall.append(float(text.split("\n")[2].split(" ")[1]))            
                f1.append(float(text.split("\n")[3].split(" ")[2])) 
                f.close()
        df.loc[i] = [exp_name] + [np.mean(precision), np.mean(recall), np.mean(f1)]
        i += 1
        
```

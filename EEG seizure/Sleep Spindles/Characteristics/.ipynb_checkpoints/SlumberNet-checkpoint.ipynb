```python
def ReadSignals():
    all_data = {}
    for file in tqdm(os.listdir(PATH)):
        if "PSG" in file:
            data = mne.io.read_raw_edf(PATH + file, verbose=False)
            raw_data = data.get_data()

            labels = np.load("data/" + file + "labels.npy")
            assert len(labels) == len(raw_data[0])
            number = int(file[8:10])
            all_data[number] = [raw_data, labels]
           
            data.close()
    return all_data
```

```python
def broad_filter(signal, fs, lowcut=0.1, highcut=35):
    """Returns filtered signal sampled at fs Hz, with a [lowcut, highcut] Hz
    bandpass."""
    # Generate butter bandpass of order 3.
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(3, (low, high), btype='band')
    # Apply filter to the signal with zero-phase.
    filtered_signal = filtfilt(b, a, signal)
    return filtered_signal
```

```python
def NormalizeAndClip(data):
    for i in tqdm(range(len(data))):
        signal = data[i]
        means = signal.mean(axis=1)[..., None]
        stds = signal.std(axis=1)[..., None]
        signal = np.clip((signal - means) / stds, a_min=-10, a_max=10)
        data[i] = signal
```

```python
def CollectAmplitude(x):
    amplitudes = np.max(x[:19], axis=1)
    return amplitudes.mean()
```

```python
def CollectFreq(x):
    centered_x = x[:19] - np.mean(x[:19], axis=1)[..., None]
    zero_cross_num = ((centered_x[:, :-1] * centered_x[:, 1:]) < 0).sum(axis=1)
    return zero_cross_num.mean() / x.shape[1]
```

```python
def CollectCharacteristics(x, labels):   
    lens = []
    ampls = []
    freqs = []
    
    c = 0
    in_event = False
    for i in range(len(labels)):
        if labels[i] == 1:
            if not in_event:
                idx_start = i
            in_event = True
            c += 1
        elif in_event:
            ampls.append(CollectAmplitude(x[:, idx_start:i]))
            freqs.append(CollectFreq(x[:, idx_start:i]))
            lens.append(c)
            c = 0
            in_event = False
                
    return np.mean(lens), np.mean(ampls), np.mean(freqs), len(lens)   
```

```python
def CreateSamples(x, labels, window=10): #window in minutes
    sample_window = 10 * 60 * 256
    inout_seq_labels = []
    inout_seq_characteristics = []
    for i in range(0, x.shape[1] - sample_window, sample_window):
        if (labels[i:i + sample_window]).max() == 1:
            train_seq = x[:, i:i+sample_window]
            train_label = labels[i:i+sample_window]
            duration, amplitude, freq, ss_number = CollectCharacteristics(train_seq, train_label)
            inout_seq_labels.append((torch.FloatTensor(train_seq), train_label))
            inout_seq_characteristics.append((torch.FloatTensor(train_seq), torch.FloatTensor([duration, amplitude, freq, ss_number])))
    return inout_seq_labels, inout_seq_characteristics
```

# Network

```python
class ResNetBlock(nn.Module):
    def __init__(self, input_dim, output_dim, kernel_size, dropout, last):
        super(ResNetBlock, self).__init__()
        
        self.dropout = nn.Dropout(dropout)
        self.relu = nn.ReLU()
        self.last = last
        
        self.conv_x = nn.Conv1d(input_dim, output_dim, kernel_size=kernel_size, padding=(kernel_size-1)//2) #p = d * (k-1) / 2
        self.bn_x = nn.BatchNorm1d(output_dim)
        
        self.conv_y = nn.Conv1d(output_dim, output_dim, kernel_size=kernel_size, padding=(kernel_size-1)//2) 
        self.bn_y = nn.BatchNorm1d(output_dim)
        
        
        self.conv_z = nn.Conv1d(output_dim, output_dim, kernel_size=kernel_size, padding=(kernel_size-1)//2)
        self.bn_z = nn.BatchNorm1d(output_dim)
        
        
        self.conv_y_sc = nn.Conv1d(input_dim, output_dim, kernel_size=kernel_size, padding=(kernel_size-1)//2)
        self.bn_y_sc = nn.BatchNorm1d(output_dim)

    def forward(self, input_tensor):
        x = self.conv_x(input_tensor)
        x = self.bn_x(x)
        x = self.dropout(x)
        x = self.relu(x)
        
      
        
        y = self.conv_y(x)
        y = self.bn_y(y)
        y = self.dropout(y)
        y = self.relu(y)
      
        
        
        z = self.conv_z(y)
        z = self.bn_z(z)
      
        
        
        if not self.last:
            sc_y = self.conv_y_sc(input_tensor)
        else:
            sc_y = input_tensor
        sc_y = self.bn_y_sc(sc_y)
      
        
        
        out = z + sc_y
        out = self.dropout(out)
        out = self.relu(out)
        return out


class SlumberNet(nn.Module):
    def __init__(self, input_dim=10*60*256, num_block=2, feature_num=27):
        super(SlumberNet, self).__init__()
        
        self.first_downsampling = nn.Conv1d(feature_num, feature_num, 20, 10, 5) #padding = (k - stride) / 2 for downsampling in stride times
        self.bn_d1 = nn.BatchNorm1d(feature_num)
        self.second_downsampling = nn.Conv1d(feature_num, feature_num, 120, 60, 30)
        self.bn_d2 = nn.BatchNorm1d(feature_num)
        
        self.relu = nn.ReLU()
        self.res_blocks = nn.ModuleList([])
        self.feature_num = feature_num
        self.factor = 32
        for i in range(num_block - 1):
            self.res_blocks.append(ResNetBlock(self.feature_num, self.factor*(2**i), 3, 0, False))
            self.feature_num = self.factor*(2**i)
        self.res_blocks.append(ResNetBlock(self.feature_num, self.factor*(2**i), 3, 0, True))
        
        self.linear = nn.Linear(256, 1)
        
    def forward(self, input_tensor):
        x = self.first_downsampling(input_tensor)

        x = self.bn_d1(x)
        x = self.relu(x)
        x = self.second_downsampling(x)
        x = self.bn_d2(x)
        x = self.relu(x)

        
        
        for i in range(len(self.res_blocks)):
            x = self.res_blocks[i](x)
            
            
        x = x.mean(axis=1) #Global Average Pooling
        out = self.linear(x)
        
        return out
        
```

```python
def CustomMSE(preds, true):
    weights = torch.FloatTensor([0.001, 1, 10, 0.1]).to('cuda')
    mse = (preds - true) ** 2
    mse *= weights
    return torch.mean(mse.sum(axis=1))
```

## Distance

```python
model_d = SlumberNet(num_block=7).to('cuda')
```

```python
def run_epochD(model, optimizer, criterion, dataloader, is_training=False):
    epoch_loss = 0

    if is_training:
        model.train()
    else:
        model.eval()

    for idx, (x, y) in tqdm(enumerate(dataloader)):
        if is_training:
            optimizer.zero_grad()
      
        #print("x", x.shape)
        
        out = model(x.to("cuda"))
        
        loss = criterion(out, y[:, 0][..., None].to("cuda"))

        if is_training:
            loss.backward()
            optimizer.step()

        epoch_loss += (loss.detach().item() / len(dataloader))


    return epoch_loss
```

```python
# define optimizer and loss function

criterion = nn.MSELoss()
#optimizer = optim.SGD(model_d.parameters(), lr = 1e-3, momentum = 0.9, weight_decay = 0.5)
optimizer = Adan(
    model_d.parameters(),
    lr = 1e-3,                  # learning rate (can be much higher than Adam, up to 5-10x)
    betas = (0.02, 0.08, 0.01), # beta 1-2-3 as described in paper - author says most sensitive to beta3 tuning
    weight_decay = 0.02         # weight decay 0.02 is optimal per author
)

#optim.Adam(model.parameters(), lr=0.001, betas=(0.9, 0.999), eps=1e-9)
scheduler = ReduceLROnPlateau(optimizer, patience = 3, factor=0.5, min_lr=0.0001)
epochs = 1000

losses_train = []
losses_test = []
best_loss = 10e9
# begin training

early_stop_count = 30
current_es = 0
for epoch in range(epochs): 
    
        
    loss_train = run_epochD(model_d, optimizer, criterion, train_dataloader, is_training=True)
    loss_val = run_epochD(model_d, optimizer, criterion, test_dataloader)
    scheduler.step(loss_val)
    losses_train.append(loss_train)
    losses_test.append(loss_val)

    if losses_test[-1] < best_loss:
        current_es = 0
        best_loss = losses_test[-1]
        torch.save(model_d.state_dict(), "./models/SlumberNet/DistanceADAN7Blocks")
    else:
        current_es += 1
    #if current_es == early_stop_count:
        #break

    clear_output(True)
    fig = plt.figure(figsize=(10, 9))

    ax_1 = fig.add_subplot(2, 1, 1)
    ax_2 = fig.add_subplot(2, 1, 2)
    ax_1.set_title('train')
    ax_1.plot(losses_train)
    ax_2.set_title('test')
    ax_2.plot(losses_test)
    plt.show()


    print('Epoch[{}/{}] | loss train:{:.6f}, test:{:.6f}'
              .format(epoch+1, epochs, loss_train, loss_val))
```

```python
def CollectPredsSingleChar(model, dataloader, i):
    model.eval()
    preds = np.empty([1])
    answers = np.empty([1])
    for idx, (x, y) in tqdm(enumerate(dataloader)):      
        
        answers = np.hstack([answers, y[:, i]])
        out = model(x)
        
        preds = np.hstack([preds, out[:, 0].cpu().detach().numpy()])
        
    return answers, preds   
```

```python
def CollectPreds(model, dataloader):
    model.eval()
    preds = np.empty([1, 4])
    answers = np.empty([1, 4])
    for idx, (x, y) in tqdm(enumerate(dataloader)):      
        #print("x", x.shape)
        answers = np.vstack([answers, y])
        out = model(x.to("cuda"))
        preds = np.vstack([preds, out.cpu().detach().numpy()])
        
    return answers, preds   
```

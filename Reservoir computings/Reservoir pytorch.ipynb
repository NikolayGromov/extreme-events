```python
class ReservoirComputings(nn.Module):
    def __init__(self, input_size=1, size=200, spectral_radius=0.001, dummy_steps=1000, alpha=0.5):
        super().__init__()
        
        self.size = size
        self.input_size  = input_size
        self.spectral_radius = spectral_radius
        self.linear = nn.Linear(self.size, self.input_size)
        self.W = self._init_reservoir_matrix()
        self.W_in = self._init_input_matrix()
        self.dummy_steps = dummy_steps
        self.alpha = alpha
        
    def _init_reservoir_matrix(self, dist_mode='uniform', diagonal_elements=0, dist_parameters=1):
        W = 0
        if dist_mode == 'uniform':
            W = np.random.rand(self.size,self.size) * 2 * dist_parameters  - dist_parameters 
            for i in range(self.size):
                W[i][i] = diagonal_elements
            E, _ = np.linalg.eig(W)
            e_max = np.max(np.abs(E))
            W /= np.abs(e_max)/self.spectral_radius
        return torch.FloatTensor(W)
        
    def _init_input_matrix(self, dist_mode='uniform', dist_parameters=1):
        W = 0
        if dist_mode == 'uniform':
            W = np.random.rand(self.size,self.input_size) * 2 * dist_parameters  - dist_parameters 
            
        return torch.FloatTensor(W)
    
    def _dummy_steps(self, time_series):
        self.r = torch.zeros(self.size)
        
        for i, x in enumerate(time_series):
            
            self.r = self.alpha * self.r + (1 - self.alpha) * nn.Tanh()(self.W @ self.r + self.W_in @ x.unsqueeze(0))
            
            

    def forward(self, x):
        
        self.r = self.alpha * self.r + (1 - self.alpha) * nn.Tanh()(self.W @ self.r + self.W_in @ x.unsqueeze(0))   
        
        
        pred_x = self.linear(self.r)
        return pred_x
                
            
                
            
```

```python
model = ReservoirComputings()
```

```python
def run_epoch(model, data, is_training=False):
    epoch_loss = 0
    data = torch.FloatTensor(data)
    if is_training:
        model.train()
        model._dummy_steps(data[:model.dummy_steps])
    else:
        model.eval()
        
    
    
    for idx, x in enumerate(data[model.dummy_steps:-1]):
        if is_training:
            optimizer.zero_grad()

        

        #x = torch.reshape(x, (64, 20, 1))
        
        y = data[model.dummy_steps + idx + 1]

        out = model(x)
        loss = criterion(out.contiguous(), y.contiguous())

        if is_training:
            loss.backward()
            optimizer.step()

        epoch_loss += (loss.detach().item())


    return epoch_loss
```
